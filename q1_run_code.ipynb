```python
from sklearn.metrics.pairwise import euclidean_distances
def find_ED(class_):
    ed_lis = []
    meanV = np.mean(class_, axis=0)
    ed = np.sqrt(np.sum(np.square(class_-meanV), axis=1))
    
    ed_lis = sorted(ed, reverse=False)
    
    #1st closest and 2nd closest
    min_1st = ed_lis[0]
    min_2nd = ed_lis[1]
    #1st furthest and 2nd furthest
    max_1st = ed_lis[-1]
    max_2nd = ed_lis[-2]
    
    #corresponding index
    min_1st_index = np.where(ed==min_1st)
    min_2nd_index = np.where(ed==min_2nd)
    max_1st_index = np.where(ed==max_1st)
    max_2nd_index = np.where(ed==max_2nd)
    return min_1st_index,min_2nd_index,max_2nd_index,max_1st_index
```

```python
#class 1
def collectImg(class_):
    meanV_1 = np.mean(class_, axis = 0)
    min_1st_idx_c1,min_2nd_idx_c1,max_2nd_idx_c1,max_1st_idx_c1 = find_ED(class_)
    close_1_c1 = class_[min_1st_idx_c1]
    close_2_c1 = class_[min_2nd_idx_c1]
    far_1_c1 = class_[max_1st_idx_c1]
    far_2_c1 = class_[max_2nd_idx_c1]

    image = np.array([meanV_1, close_1_c1, close_2_c1, far_2_c1, far_1_c1])
    index = [min_1st_idx_c1,min_2nd_idx_c1,max_2nd_idx_c1,max_1st_idx_c1]
    return [image,index]
```

```python
# Q1.6
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA

RMSE = []
for k in [5,20,50,200]:
    pca = PCA(n_components=k,random_state=1)
    pca = pca.fit(Xtrn_nm)

    for i in range(10):
        #first sample of a single class
        ori = Xtrn_nm[Ytrn==i,:][0]
        ori = ori.reshape((1,Xtrn_nm[Ytrn == i][0].shape[0]))
        
        Xtrn_class = pca.transform(ori)
        Xtrn_class = np.dot(Xtrn_class, pca.components_)
    
        mse = mean_squared_error(Xtrn_class, ori)
        rmse = np.sqrt(mse)
        rmse = "K = " + str(k) + "," + "class = " + str(i) + ", " + "RMSE = " + str(rmse)
        RMSE.append(rmse)
        
print(RMSE)
```
